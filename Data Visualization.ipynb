```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(nrows=4,ncols=1,figsize=(6,8))
axes = [ax1,ax2,ax3,ax4]
location = ['top','bottom','left','right']
for ax in axes:
    ax.plot(death_toll['Month'],death_toll['New_deaths'],alpha = 0.1,color='#af0b1e')
    ax.set_yticklabels([])
    #ax.set_xticklabels([])
    ax.tick_params(bottom=False, left=False)
    for loc in location:
        ax.spines[loc].set_visible(False)
ax1.plot(death_toll['Month'][:3],death_toll['New_deaths'][:3],linewidth=2.5,color='#af0b1e')
ax2.plot(death_toll['Month'][2:6],death_toll['New_deaths'][2:6],linewidth=2.5,color='#af0b1e')
ax3.plot(death_toll['Month'][5:9],death_toll['New_deaths'][5:9],linewidth=2.5,color='#af0b1e')
ax4.plot(death_toll['Month'][8:],death_toll['New_deaths'][8:],linewidth=2.5,color='#af0b1e')

# add labels
ax1.text(0.5,-80,'0', alpha=0.5)
ax1.text(3.5,2000,'1,844', alpha=0.5)
ax1.text(11.5,2400,'2,247', alpha=0.5)

# More labels months 
ax1.text(1.1,-300,'Jan - Mar', color='#af0b1e', weight = 'bold', rotation=3)
ax2.text(3.7, 800,'Mar - Jun', color='#af0b1e', weight = 'bold')
ax3.text(7.1, 500,'Jun - Oct', color='#af0b1e', weight = 'bold')
ax4.text(10.5,600,'Oct - Dec', color='#af0b1e', weight = 'bold', rotation=45)

# add title 
ax1.text(0.5,3500,'The virus kills 851 people each day',size=14,weight = 'bold')
ax1.text(0.5,3150,'Average number of daily deaths per month in the US', size=12)

deaths = [2398,126203,227178,295406]
proportions = [round(death/295406,2) for death in deaths]
xmax_vals = [round(0.5  +proportion *0.3 ,3) for proportion in proportions]
            

for ax,xmax,death in zip(axes,xmax_vals,deaths):
    ax.axhline(y=1600,xmin=0.5,xmax=0.8, linewidth=6, color='#af0b1e', alpha=0.1)
    ax.axhline(y=1600,xmin=0.5,xmax=xmax, linewidth=6, color='#af0b1e')
    ax.text(7.5, 1850,format(death,','), color='#af0b1e', weight = 'bold')
plt.show()
```

```python
red_corr = red_wine.corr()['quality'][:-1]
print(red_corr)
```

```python
white_corr = white_wine.corr()['quality'][:-1]
print(white_corr)
```

```python
positive_white = white_corr>0
color_map_white = positive_white.map({True:'#33A1C9', False:'#ffae42'})
color_map_white

positive_red = red_corr>0
color_map_red = positive_red.map({True:'#33A1C9', False:'#ffae42'})
color_map_red
```

```python
style.use('fivethirtyeight')
fig,ax = plt.subplots(figsize=(9,5))
ax.barh(white_corr.index,white_corr, left=2, height = 0.5, color=color_map_white)
ax.barh(red_corr.index,red_corr, height =0.5, left=-0.1,color=color_map_red)
ax.grid(b=False)
ax.set_xticklabels([])
ax.set_yticklabels([])

x_coords = {'Alcohol': 0.82, 'Sulphates': 0.77, 'pH': 0.91,
            'Density': 0.80, 'Total Sulfur Dioxide': 0.59,
            'Free Sulfur Dioxide': 0.6, 'Chlorides': 0.77,
            'Residual Sugar': 0.67, 'Citric Acid': 0.76,
            'Volatile Acidity': 0.67, 'Fixed Acidity': 0.71}
y_coord = 9.8
for y_label, x_coord in x_coords.items():
    ax.text(x_coord, y_coord, y_label)
    y_coord -=1
    
ax.axvline(0.5, color='grey', alpha=0.1, lw=1, ymin=0.1, ymax=0.9)
ax.axvline(1.45, color='grey', alpha=0.1, lw=1, ymin=0.1, ymax=0.9)
ax.axhline(-1, c='grey', alpha=0.5, lw=1, xmin=0.01, xmax=0.32)
ax.axhline(-1, c='grey', alpha=0.5, lw=1, xmin=0.67, xmax=0.98)

ax.text(-0.7,-1.7,'-0.5' + ' '*31 +'+0.5',c='grey', alpha=0.5)
ax.text(1.43,-1.7,'-0.5' + ' '*31 +'+0.5',c='grey', alpha=0.5)

ax.axhline(11, c='grey', alpha=0.5, lw=1, xmin=0.01, xmax=0.32)
ax.axhline(11, c='grey', alpha=0.5, lw=1, xmin=0.67, xmax=0.98)
ax.text(-0.33,11.2,'RED WINE', alpha=0.5, weight='bold')
ax.text(1.75,11.2,'WHITE WINE', alpha=0.5, weight='bold')


ax.text(-0.7, -2.9,
        '©DATAQUEST'+' '*94 +'Source: P. Cortez et al.',
        color = '#f0f0f0',
        backgroundcolor = '#4d4d4d',
        size=12)

ax.text(-0.7,13.5,'Wine Quality Most Strongly Correlated With Alcohol Level', weight='bold',size=17)
ax.text(-0.7,12.7,'Correlation values between wine quality and wine properties (alcohol, pH, etc.)')


plt.show()
```
