# Data Cleaning

```python
exchange_rates.rename(columns={'[US dollar ]':'US_dollar','Period\\Unit:':'Time'}, inplace=True)
exchange_rates.head()
```

```python
exchange_rates['Time']=pd.to_datetime(exchange_rates['Time'])# change the Time column to dateTime data type
exchange_rates.sort_values('Time', inplace=True) # Sort the values by time in ascending order
exchange_rates.reset_index(drop=True, inplace=True) # reset the index and drop the initial index
```

```python
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-'] # drop all the rows where '-' appears
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
# Rolling mean - 30 days
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar.head(60)
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```
